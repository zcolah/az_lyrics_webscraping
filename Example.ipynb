Here is a demonstration of the azlyrics library. It has the following functions:

    - get_songs(artist_name)
    - get_lyrics_for_artist(artist_name)
    - get_lyrics_details(artist_name)
    - process_song_lyrics(lyrics) where lyrics is a list returned by get_lyrics_for_artist

Other functions which do exist are: 

    - get_song_links(song_list_link) where song_list_link is the link of an artist on azlyrics.com
    - lyric_counter(lyrics) where lyrics is a list returned by get_lyrics_for_artist

```python
# Import your functions
import azlyrics as l

# Get songs
artist = 'Zane'
all_songs = l.get_songs(artist)

# Get lyrics -- internally, you may want to another function in your library, such as get_lyrics()
lyrics = l.get_lyrics_for_artist(artist)

# Save data -- this function should identify the name of the artist and use it in the .csv file it saves

# Process lyrics

lyrics_details = l.get_lyrics_details(artist)
lyrics_details.to_csv('zane.csv')

# you can also do
processed_data = l.process_song_lyrics(lyrics)
processed_data.to_csv('zane_type2.csv')
```
