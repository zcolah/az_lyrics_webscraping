```python
# Returns the artist's page which list's their songs
# Returns the artist's page which list's their songs
def get_artist_page(artist_name):
    
    # Import beautiful soup
    from bs4 import BeautifulSoup as bs
    # query = "https://search.azlyrics.com/search.php?q="+artist_name
    query = "https://www.lyrics.com/serp.php?st=" + artist_name + "&qtype=2"
    
    # Use the `get` method of the requests library to fetch the page content
    import requests as r
    page = r.get(query)
    
    # Use bs to parse the HTML returned
    soup = bs(page.content, "html.parser")
    
    # directory location 
    dir_locs = soup.find_all("a", string=artist_name)
    dir_loc = (dir_locs[0])
    dir_link = dir_loc['href']
    print(dir_link)
    artist = dir_link[7:].split("/")
    
    artist_page_link = "https://www.lyrics.com/artist.php?name=" + artist[0] + "&aid="+artist[1]+"&o=1"
    
    # returns the first search result for the artist_name
    return artist_page_link
```

```python
# Returns the links of all the songs of the artist
def get_song_links(song_list_link):
    # Import beautiful soup
    from bs4 import BeautifulSoup as bs
    query = song_list_link
    # Use the `get` method of the requests library to fetch the page content
    import requests as r
    page = r.get(query)
    soup = bs(page.content, "html.parser")
    
    song_tags = soup.select('td[class="tal qx"]')
    #["https://www.azlyrics.com"+ (s.get('href'))[2:] for s in song_tags]
    
    for song in song_tags:
        song_link = type(song.strong)
        print(song_link)
    
    return len(song_tags)
```
